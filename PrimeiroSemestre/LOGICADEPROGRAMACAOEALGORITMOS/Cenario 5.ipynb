```
total = 0
while True:
  valor = int(input('Informe o valor a ser somado'))
  if valor == 0:
    print('valor total da soma eh {}'.format(total))
    break
  elif valor > 0 and valor < 10:
    total+= valor #total = total + valor
  else:
    print ('Valor informado incorreto, favor informar valor valido')
    
```
